```python
import os
import cv2
from ultralytics import YOLO
from ultralytics.utils.plotting import colors
from base64 import b64encode
from IPython.display import HTML

# Absolute paths for model and video
model_path = os.path.abspath("/content/drive/MyDrive/sesan/models_for_laparoscopic_videos/best.pt")
video_path = os.path.abspath("/content/drive/MyDrive/sesan/videos/video002.mp4")

# Check if the model file exists
if not os.path.isfile(model_path):
    raise FileNotFoundError(f"Model file not found: {model_path}")

# Check if the video file exists
if not os.path.isfile(video_path):
    raise FileNotFoundError(f"Video file not found: {video_path}")

# Load the model
model = YOLO(model_path)  # segmentation model
names = model.model.names

# Load the video
cap = cv2.VideoCapture(video_path)
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

# Check if video is loaded successfully
if not cap.isOpened():
    raise Exception("Error: Could not open video.")

# Define the codec and create VideoWriter object for MP4 format
output_path_mp4 = "/content/drive/MyDrive/sesan/result_videos/video0002.mp4"
out = cv2.VideoWriter(output_path_mp4, cv2.VideoWriter_fourcc(*"mp4v"), fps, (w, h))

while True:
    ret, im0 = cap.read()
    if not ret:
        print("Video frame is empty or video processing has been successfully completed.")
        break

    # Run inference
    results = model.predict(im0)

    if results[0].masks is not None:
        clss = results[0].boxes.cls.cpu().tolist()
        masks = results[0].masks.xy
        for mask, cls in zip(masks, clss):
            color = colors(int(cls), True)
            pts = mask.reshape((-1, 1, 2)).astype(int)
            cv2.polylines(im0, [pts], isClosed=True, color=color, thickness=2)
            cv2.putText(im0, names[int(cls)], (pts[0][0][0], pts[0][0][1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)

    # Write the frame to the output video
    out.write(im0)

# Release resources
out.release()
cap.release()

# Check if MP4 file was created successfully
if not os.path.isfile(output_path_mp4):
    raise FileNotFoundError(f"MP4 file not created: {output_path_mp4}")

# Show the resulting video
def show_video(file_name, width=640):
    mp4 = open(file_name, 'rb').read()
    data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
    return HTML("""
    <video width="{0}" controls>
        <source src="{1}" type="video/mp4">
    </video>
    """.format(width, data_url))

# Display the output video
show_video(output_path_mp4, width=960)
```

```python
import os
import cv2
from ultralytics import YOLO
from ultralytics.utils.plotting import colors
from shapely.geometry import Polygon
from shapely.validation import explain_validity
from base64 import b64encode
from IPython.display import HTML

# Absolute paths for model and video
model_path = os.path.abspath("best.pt")
video_path = os.path.abspath("videos/video002.mp4")

# Check if the model file exists
if not os.path.isfile(model_path):
    raise FileNotFoundError(f"Model file not found: {model_path}")

# Check if the video file exists
if not os.path.isfile(video_path):
    raise FileNotFoundError(f"Video file not found: {video_path}")

# Load the model
model = YOLO(model_path)  # segmentation model
names = model.model.names

# Load the video
cap = cv2.VideoCapture(video_path)
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

# Check if video is loaded successfully
if not cap.isOpened():
    raise Exception("Error: Could not open video.")

# Define the codec and create VideoWriter object for MP4 format
output_path_mp4 = "buffer_30secs.mp4"
out = cv2.VideoWriter(output_path_mp4, cv2.VideoWriter_fourcc(*"mp4v"), fps, (w, h))

# Function to check for intersections between tool and liver polygons
def check_intersection(liver_polygons, tool_polygons, buffer_distance=10):
    for liver_poly in liver_polygons:
        buffered_liver_poly = liver_poly.buffer(buffer_distance)  # Create buffer zone around the liver
        for tool_poly in tool_polygons:
            if not buffered_liver_poly.is_valid or not tool_poly.is_valid:
                print(f"Invalid polygons detected: {explain_validity(buffered_liver_poly)}, {explain_validity(tool_poly)}")
                continue
            if buffered_liver_poly.intersects(tool_poly):
                return True
    return False

# Function to create a Shapely polygon from mask points
def create_polygon_from_mask(mask_points):
    return Polygon(mask_points)

while True:
    ret, im0 = cap.read()
    if not ret:
        print("Video frame is empty or video processing has been successfully completed.")
        break

    # Run inference
    results = model.predict(im0)

    if results[0].masks is not None:
        clss = results[0].boxes.cls.cpu().tolist()
        masks = results[0].masks.xy

        liver_polygons = []
        tool_polygons = []

        for mask, cls in zip(masks, clss):
            color = colors(int(cls), True)
            pts = mask.reshape((-1, 1, 2)).astype(int)
            cv2.polylines(im0, [pts], isClosed=True, color=color, thickness=2)
            cv2.putText(im0, names[int(cls)], (pts[0][0][0], pts[0][0][1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)

            # Create polygons for liver and tools
            polygon = create_polygon_from_mask(mask)
            if names[int(cls)] == 'gallbladder':
                liver_polygons.append(polygon)
            elif names[int(cls)] == 'tool':
                tool_polygons.append(polygon)

        # Check for intersections and draw alert if necessary
        if check_intersection(liver_polygons, tool_polygons, buffer_distance=10):  # Adjust buffer distance as needed
            cv2.putText(im0, 'Warning: Tool is near the gallbladder!', (10, 30),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2, cv2.LINE_AA)

    # Write the frame to the output video
    out.write(im0)

# Release resources
out.release()
cap.release()

# Check if MP4 file was created successfully
if not os.path.isfile(output_path_mp4):
    raise FileNotFoundError(f"MP4 file not created: {output_path_mp4}")

# Show the resulting video
def show_video(file_name, width=640):
    mp4 = open(file_name, 'rb').read()
    data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
    return HTML("""
    <video width="{0}" controls>
        <source src="{1}" type="video/mp4">
    </video>
    """.format(width, data_url))

# Display the output video
show_video(output_path_mp4, width=960)
```

```python
import os
import cv2
from ultralytics import YOLO
from ultralytics.utils.plotting import colors
from shapely.geometry import Polygon
from shapely.validation import explain_validity
from pydub import AudioSegment
from pydub.playback import play

# Load the alert sound
alert_sound = AudioSegment.from_wav("alert.wav")

# Function to play the alert sound
def play_alert_sound():
    play(alert_sound)

# Load the YOLO model
model = YOLO("best.pt")  # Ensure the model is in the same directory

# Names of the classes in the model
names = model.model.names

# Access the laparoscopic camera feed
cap = cv2.VideoCapture(0)  # Use appropriate device index for your camera

# Check if camera is opened successfully
if not cap.isOpened():
    raise Exception("Error: Could not access the laparoscopic camera.")

# Function to check for proximity between tool and gallbladder polygons
def check_proximity(gallbladder_polygons, tool_polygons, threshold=10.0):
    for gallbladder_poly in gallbladder_polygons:
        for tool_poly in tool_polygons:
            if not gallbladder_poly.is_valid or not tool_poly.is_valid:
                print(f"Invalid polygons detected: {explain_validity(gallbladder_poly)}, {explain_validity(tool_poly)}")
                continue
            if gallbladder_poly.distance(tool_poly) < threshold:
                return True
    return False

# Function to create a Shapely polygon from mask points
def create_polygon_from_mask(mask_points):
    return Polygon(mask_points)

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame.")
        break

    # Run inference
    results = model.predict(frame)

    if results[0].masks is not None:
        clss = results[0].boxes.cls.cpu().tolist()
        masks = results[0].masks.xy

        gallbladder_polygons = []
        tool_polygons = []

        for mask, cls in zip(masks, clss):
            color = colors(int(cls), True)
            pts = mask.reshape((-1, 1, 2)).astype(int)
            cv2.polylines(frame, [pts], isClosed=True, color=color, thickness=2)
            cv2.putText(frame, names[int(cls)], (pts[0][0][0], pts[0][0][1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)

            # Create polygons for gallbladder and tools
            polygon = create_polygon_from_mask(mask)
            if names[int(cls)] == 'gallbladder':
                gallbladder_polygons.append(polygon)
            elif names[int(cls)] == 'tool':
                tool_polygons.append(polygon)

        # Check for proximity and trigger alert if necessary
        if check_proximity(gallbladder_polygons, tool_polygons):
            cv2.putText(frame, 'Danger: Tool is near the gallbladder!', (10, 30),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2, cv2.LINE_AA)
            play_alert_sound()

    # Display the frame
    cv2.imshow("Laparoscopic Feed", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```

```python
import torch
from ultralytics import YOLO  # Import the YOLO class from the ultralytics package
import yaml
import os

def main():
    # Verify PyTorch version and that CUDA is available
    print("PyTorch version:", torch.__version__)
    print("CUDA available:", torch.cuda.is_available())

    # Set the device to GPU if available; otherwise, fall back to CPU
    device_str = "cuda" if torch.cuda.is_available() else "cpu"
    device = torch.device(device_str)
    print(f"Selected device: {device_str}")

    # Print the current working directory
    current_directory = os.getcwd()
    print("Current working directory:", current_directory)

    # Path to your data.yaml file (assumed relative to current directory)
    data_yaml_path = os.path.join(current_directory, '/content/drive/MyDrive/3D_types_of_tools/tools_3d-1/data.yaml')

    # Verify the paths inside data.yaml
    try:
        with open(data_yaml_path, 'r') as file:
            data_yaml = yaml.safe_load(file)

        train_path = os.path.join(current_directory, data_yaml.get('train', ''))
        val_path = os.path.join(current_directory, data_yaml.get('val', ''))

        print(f"Train images path: {train_path}")
        print(f"Validation images path: {val_path}")

        # Check if the paths exist
        if not os.path.exists(train_path):
            print(f"Error: Training path does not exist: {train_path}")
        else:
            print(f"Training path verified: {train_path}")

        if not os.path.exists(val_path):
            print(f"Error: Validation path does not exist: {val_path}")
        else:
            print(f"Validation path verified: {val_path}")

        # Load the YOLOv8 model for instance segmentation and keep it on GPU/CPU
        try:
            model = YOLO('yolov8s-seg.pt').to(device)  # Load the pre-trained YOLOv8 model for segmentation

            # Train the model
            model.train(
                data=data_yaml_path,  # Path to the data.yaml file
                epochs=10,  # Number of training epochs
                imgsz=640,  # Image size for training
                batch=32,  # Batch size
                lr0=0.001,  # Learning rate
                save_period=1,  # Save checkpoints every epoch
                workers=4,  # Number of workers for data loading
                project='runs/train',  # Directory to save results
                name='yolov8_instance_segmentation',  # Name of the training run
                device=device_str,  # Specify the device as 'cuda' or 'cpu'
                task='segment'  # Specify the task as 'segment' for instance segmentation
            )
            print("Training completed successfully.")
        except Exception as train_exception:
            print("An error occurred during training:", train_exception)

    except Exception as yaml_exception:
        print("An error occurred while reading the YAML file:", yaml_exception)

if __name__ == '__main__':
    from multiprocessing import freeze_support
    freeze_support()  # Only needed if freezing the script into an executable
    main()
```
